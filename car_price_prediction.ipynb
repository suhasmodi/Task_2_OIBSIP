```python
data['avgmpg']=(data['citympg']+data['highwaympg'])/2
data=data.drop(['citympg','highwaympg'],axis=1)
```

```python
val_col=data[['symboling','doornumber','aspiration','carbody','drivewheel','fuelsystem','enginetype','cylindernumber','fueltype','enginelocation']]
for column in val_col:
  print(column,":")
  print(data[column].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in val_col:
  data[column]=le.fit_transform(data[column])

#two:1 , four:0
#gas:1 , diesel:0
```

```python
y=data['price']
# val_col=data[['symboling','doornumber','aspiration','carbody','drivewheel','fuelsystem','enginetype','cylindernumber','fueltype']]

x=data.drop(['car_ID','price','drivewheel','doornumber','CarName','enginelocation','carheight','stroke','compressionratio','avgmpg'],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score,r2_score
from sklearn.metrics import mean_absolute_percentage_error,mean_squared_error


models=[LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor()]
```

```python
for model in models:
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)
  print(f'{model}:')
  print("r2_score: ",r2_score(y_pred,y_test))
  print(model.score(x_test, y_pred))
  print(mean_absolute_percentage_error(y_test,y_pred)*100)
  print(mean_squared_error(y_test,y_pred))
```

```python
model=RandomForestRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(f'{model}:')
print("r2_score: ",r2_score(y_pred,y_test))
print("score: ",model.score(x_test, y_pred))
print("MAPE",mean_absolute_percentage_error(y_test,y_pred)*100)
```
